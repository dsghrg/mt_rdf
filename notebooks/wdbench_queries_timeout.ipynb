```python
df_paths_res = pd.read_csv('results/wdbench/timeout_api/results_paths.csv')

df_c2rpqs_res = pd.read_csv('results/wdbench/timeout_api/results_c2rpqs.csv')
df_multiple_bgps_res = pd.read_csv('results/wdbench/timeout_api/results_multiple_bgps.csv')
df_opts_res = pd.read_csv('results/wdbench/timeout_api/results_opts.csv')
df_single_bgps_res = pd.read_csv('results/wdbench/timeout_api/results_single_bgps.csv')
```

```python
df_paths_res.columns
```

```python
def prep_res_df(df, qtype):
    df.drop(['Unnamed: 0'], axis=1, inplace=True)
    df['q_type'] = qtype
    df.rename(columns={'query_id':'q_id'}, inplace=True)
    return df
```

```python
df_paths_res = prep_res_df(df_paths_res, QueryType.paths.value)
df_c2rpqs_res = prep_res_df(df_c2rpqs_res, QueryType.c2rpqs.value)
df_multiple_bgps_res = prep_res_df(df_multiple_bgps_res, QueryType.multiple_bgps.value)
df_opts_res = prep_res_df(df_opts_res, QueryType.opts.value)
df_single_bgps_res = prep_res_df(df_single_bgps_res, QueryType.single_bgps.value)
```

```python
df_all_res = pd.concat([df_paths_res, df_c2rpqs_res, df_multiple_bgps_res, df_opts_res, df_single_bgps_res])
```

```python
df_all_res
```

```python
df_all_timeout = df_all_res[df_all_res['exec_time'] == 0]
df_all_timeout
```

```python
df_stats_q_time = pd.merge(wdbench.queries, df_all_timeout, on=['q_type', 'q_id'])
```
