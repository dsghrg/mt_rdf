```python
fig, axs = plt.subplots(3, 2)
fig.set_size_inches(10, 10)

axs[0, 0].boxplot(df_paths['exec_time'])
axs[0, 0].set_title('paths')
axs[0, 0].set_ylabel('exectuion time in s')
axs[0, 1].boxplot(df_c2rpqs['exec_time'])
axs[0, 1].set_title('c2rpqs')
axs[0, 1].set_ylabel('exectuion time in s')
axs[1, 0].boxplot(df_multiple_bgps['exec_time'])
axs[1, 0].set_title('multiple_bgps')
axs[1, 0].set_ylabel('exectuion time in s')
axs[1, 1].boxplot(df_opts['exec_time'])
axs[1, 1].set_title('opts')
axs[1, 1].set_ylabel('exectuion time in s')
axs[2, 1].boxplot(df_single_bgp['exec_time'])
axs[2, 1].set_title('single_bgp')
axs[2, 1].set_ylabel('exectuion time in s')

fig.subplots_adjust(left=0.08, right=0.98, bottom=0.05, top=0.9,
                    hspace=0.4, wspace=0.3)

plt.show()
```
